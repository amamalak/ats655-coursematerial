```
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy import stats, odr

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = (12.0/2, 8.0/2)
#%matplotlib inline    
#import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```
# Data that works
N = 500
X = np.random.uniform(low=0,high=4,size=(N,))
Y = 4*X-3+np.random.normal(0,4,size=(N,))

plt.scatter(X, Y)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```
# Data that doesn't work

# filename = '/Users/eabarnes/GoogleDrive/WORK/TEACHING/ATS655/shared/python_code/jupyter_notebooks/data/gradDescent_data.csv'
# data = pd.read_csv(filename)

url = 'https://github.com/chasinginfinity/ml-from-scratch/raw/master/02%20Linear%20Regression%20using%20Gradient%20Descent/data.csv'
data = pd.read_csv(url)

X = np.array(data.iloc[:, 0])
Y = np.array(data.iloc[:, 1])

X = (X-np.mean(X))/np.std(X)
#Y = Y-np.mean(Y)

plt.scatter(X, Y)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```
# separate training and testing data
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.2)

plt.scatter(Xtrain, Ytrain, color='cornflowerblue', label='training data')
plt.scatter(Xtest, Ytest, color='fuchsia', label = 'testing data')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend()
plt.show()
```

```
# define functional fits
def my_fit(m,c,x):
    # m is the slope of the line
    # c is the y-intercept
    # x is the values to evaluate
    # output: y, the evaluated values
    y = m*x + c
    return y

def my_gradLoss(xtrain,ytrain,ypred):
    n = float(len(xtrain))
    gradLoss_m = -2/n * sum(xtrain * (ytrain - ypred)) #derivative wrt m
    gradLoss_c = -2/n * sum((ytrain - ypred)) #derivative wrt c
    
    return gradLoss_m, gradLoss_c
```

```
# Building the model
m = np.random.uniform(low=-1,high=1)
c = np.random.uniform(low=-1,high=1)
print (m, c)

L = 0.001  # The learning Rate
epochs = 10000  # The number of iterations to perform gradient descent

print('  slope (m)             y-int (c)')
print('----------------------------------')

errorHistory = np.empty((epochs,))
# Performing Gradient Descent 
for i in range(epochs): 
    
    Y_pred = my_fit(m,c,Xtrain)  # The current predicted value of Y
    gradLoss_m, gradLoss_c = my_gradLoss(Xtrain, Ytrain, Y_pred)
    
    m = m - L * gradLoss_m  # Update m
    c = c - L * gradLoss_c  # Update c
    
    errorHistory[i] = 1/float(len(Y))*np.sum((Ytrain - Y_pred)**2)
    
    #print (m, c)
print('done training')    
print (m, c)
```

```
a1 = np.cov(Xtrain,Ytrain)[0,1]/np.var(Xtrain)
a0 = np.mean(Ytrain) - a1*np.mean(Xtrain)
print(a1,a0)
```

```
# Making predictions - FINAL
Y_pred = my_fit(m,c,Xtrain)
Y_predTest = my_fit(m,c,Xtest)

plt.scatter(X, Y, color='black', label = 'data') 
plt.plot(Xtrain, Y_pred, 'x', color='cornflowerblue', label = 'training data') 
plt.plot(Xtest, Y_predTest, 'x', color='fuchsia', label= 'testing data') 

plt.plot(Xtrain, Y_pred,'-k', label='fit', linewidth=3) # regression line

slope, intercept, r_value, p_value, std_err = stats.linregress(np.squeeze(X),np.squeeze(Y))
plt.plot(Xtrain,intercept+Xtrain*slope,'-',color = 'orange', label = 'TRUTH - LSQ: x vs y', linewidth=1)
plt.plot(Xtrain,a0+Xtrain*a1,'--',color = 'green', label = 'TRUTH - THEORY', linewidth=1)
print(slope, intercept)

plt.legend()
plt.show()
```

```

#print the error history
plt.figure()
plt.plot(np.arange(0,len(errorHistory)),errorHistory,'.-')
plt.title('loss function')
#plt.ylim(0,100)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

```
# define functional fits
def my_fit(a,x):
    # m is the slope of the line
    # c is the y-intercept
    # x is the values to evaluate
    # output: y, the evaluated values
    y = a[0] + a[1]*x + a[2]*(x**2) + a[3]*(x**3) + a[4]*(x**4) + a[5]*(x**5)
    return y

def my_gradLoss(xtrain,ytrain,ypred):
    n = float(len(xtrain))
    
    gradLoss_a = np.empty(6,)
    gradLoss_a[0] = -2/n * sum((ytrain - ypred)) #derivative wrt a0
    gradLoss_a[1] = -2/n * sum(xtrain * (ytrain - ypred)) #derivative wrt a1
    gradLoss_a[2] = -2/n * sum((xtrain**2) * (ytrain - ypred)) #derivative wrt a2
    gradLoss_a[3] = -2/n * sum((xtrain**3) * (ytrain - ypred)) #derivative wrt a3
    gradLoss_a[4] = -2/n * sum((xtrain**4) * (ytrain - ypred)) #derivative wrt a4
    gradLoss_a[5] = -2/n * sum((xtrain**5) * (ytrain - ypred)) #derivative wrt a5    
    
    return gradLoss_a
```

```
# make data
a_truth = -1.,-6.,5.,5.,-5,1. # vector of coefficients
x = np.random.uniform(low=-1.1,high=3.1,size=200)
y = my_fit(a_truth,x)

plt.figure()
plt.plot(x,y,'ok')
plt.title('data')
plt.show()
```
